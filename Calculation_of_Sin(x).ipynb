## The purpose is to develop a model for the calculation of Sin(x)

```python
import numpy as np
import tensorflow as tf
```

```python
x=np.linspace(-2*np.pi,2*np.pi,1000).reshape(-1,1)
y=np.sin(x)
```

### ANN Model

```python
ann=tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units=40,activation='sigmoid'))
```

```python
ann.add(tf.keras.layers.Dense(units=12,activation='sigmoid'))
```

```python
ann.add(tf.keras.layers.Dense(units=1))
```

```python
ann.compile(loss='mean_squared_error',optimizer='adam',metrics=['mean_squared_error'])
```

```python
ann.fit(x,y,epochs=1000,batch_size=16)
```

```python
ann.predict([-90*np.pi/180])
```

```python
tf.keras.utils.plot_model(ann, show_shapes='true')
```
